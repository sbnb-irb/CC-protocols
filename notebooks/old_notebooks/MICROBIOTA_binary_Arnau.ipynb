## Signature 0: data standardization and (optional) sanitization

```python
dataset = 'M1.001'
```

```python
# Instantiation of signature 0 data structures for the new space: full and reference
# signature(self, dataset, cctype)
sign0 = cc_local.signature(dataset,'sign0')
# Before each new signature fitting step it is important to clean up the corresponding space directory: this is both crucial and dangerous!
# It is essential to remove all files, related to the space signature, which have been written from previous runs to avoid conflicts.
# At the same time, this cleaning step will make a fresh start for the signature so any file which would be important to retain must be moved (saved) in advance.

# Cleaning both full and reference datasets
sign0.clear_all()

# Creating the signature model
# We produce a sign0 full and a sign0 reference. Data are sorted (keys and features).
# A signature which has been redundancy-reduced ('reference`) has fewer molecules. The keys are molecules from the `full` signature and values are molecules from the`reference` set.

# Parameters: see for further details http://packages.sbnb-pages.irbbarcelona.org/chemical_checker/_autosummary/chemicalchecker.core.sign0.sign0.html
#    X (matrix or file): actual data.
#    keys (array): row names.
#    features (array): column names.
#    do_triplets (boolean): set to False to avoid drawing triplets from the CC
#    overwrite (boolean): overwriting the existing signature
#    sanitize (boolean): set to True to sanitize the dataset with trimming of less informative rows/columns. Setting 'sanitizer_kwargs' parameters different from the default ones is possible. 
#    validations (boolean): set to False to skip post-signature data validations (including diagnostics which is run separately in a later step)
#                           Current validation tests are performed on MOA (Mode Of Action) and ATC (Anatomical Therapeutic Chemical) corresponding to B1.001 and E1.001 space datasets.
sign0.fit(X=df.values, keys=list(df.index), features=list(df.columns), do_triplets=False, overwrite=True, sanitize=True, validations=False)
print(sign0.shape)
```

# Signature 1: Compression of the signatures with a dimensionality typically retaining 90% of the original variance

```python
# signature 0 model is used to fit signature 1: 
# Sign1 keep most of the complexity of the original data and can be used for similarity calculations.
sign0 = cc_local.signature(dataset, 'sign0')
# Instantiation of signature 1 data structures: full and reference 
sign1 = cc_local.signature(dataset, 'sign1')
# cleaning up sign1 full and sign1 reference file directories: if just one of the two signatures (full or reference) has to be clean use signX.clear()
sign1.clear_all()

# Creating the signature model: the typical preprocessing is a PCA (continuous data) or TF-IDF LSI. 
# Parameters: see for further details http://packages.sbnb-pages.irbbarcelona.org/chemical_checker/_autosummary/chemicalchecker.core.sign1.sign1.html 
#    sign0 (signature data structure): a signature 0 
#    overwrite (boolean): overwriting the existing signature 
#    validations (boolean): set to False to skip post-signature data validations (including diagnostics which is run separately in a later step) 
#                           Current validation tests are performed on MOA (Mode Of Action) and ATC (Anatomical Therapeutic Chemical) corresponding to B1.001 and E1.001 space datasets. 
 #    model parameters different from the default ones can be passed to the fitting method (i.e. scale_kwargs={}, pca_kwargs={}, lsi_kwargs={}) 
sign1.fit(sign0, overwrite=True, validations=False, metric_learning=False)
print(sign1.shape)
```

## Signature2: Network embedding of the similarity matrix derived from signatures 1 

```python
sign2 = cc_local.signature(dataset,'sign2')
# cleaning up sign2 full and sign2 reference file directories: if just one of the two signatures (full or reference) has to be clean use signX.clear()
sign2.clear_all()

# In this step of the pipeline is convenient, for computation, to work with the non-redundant set of signatures (reference). 
# At the end a mapping of the reference to the full set of molecules is performed. 
sign2_ref = cc_local.get_signature('sign2','reference',dataset)
sign1 = cc_local.get_signature('sign1', 'full', dataset)
neig1_ref = cc_local.get_signature('neig1', 'reference', dataset)
neig1 = cc_local.get_signature('neig1', 'full', dataset)
# Signature 2 are network embeddings with fixed length and capture both explicit and implicit similarity relationships in the data.
# These signatures are the result of a two-step process:
#   Transforming nearest-neighbor of signature type 1 to graph.
#   Performing network embedding with Node2Vec.

# signature 2 given signature 1 and its nearest neighbors
# oos_predictor (boolean): set to False for skipping the predictor on out-of-space compounds 
sign2_ref.fit(sign1,neig1, oos_predictor=False, overwrite=True, validations=False)
```

```python
# initialization of sign3
sign3 = cc_local.signature(dataset, 'sign3')
# cleaning up sign3 full and sign3 reference file directories: if just one of the two signatures (full or reference) has to be clean use signX.clear()
sign3.clear_all()
```

```python
# # Create a list of sign2 to feed sign3: 
# using the 25 CC spaces + the new custom space
sign2_list = list()
# 25 CC space
for ds in cc_local.coordinates:
    ds += '.001'
    sign2_list.append(cc_local.get_signature('sign2', 'full', ds))
# adding new space to CC
sign2_list.append(cc_local.get_signature('sign2','full',dataset))
len(sign2_list)
```

```python
sign2_self = cc_local.signature(dataset, 'sign2')
sign1_self = cc_local.signature(dataset, 'sign1')
```

```python
# Creating the signature model: Signatures type 3 are available for all molecules of the ChemicalChecker universe (~1M molecules)  
# Using HPC to run the model fit: more than 1CPU are requested reducing drastically the computation time because of parallelization 

# Parameters: see for further details http://packages.sbnb-pages.irbbarcelona.org/chemical_checker/_autosummary/chemicalchecker.core.sign3.sign3.html 
#    sign2_list (list of signatures): list of sign2 to learn from 
#    sign2_self (signature): sign2 of the current space (on which the sign3 fitting is run) 
#    sign1_self (signature): sign1 of the current space (on which the sign3 fitting is run) 
#    sign2_universe(string): path to the union of signatures 2 for all molecules in the CC universe (computed once only) 
#    complete_universe (string): add chemistry information for molecules not in the universe. 
#                                ‘full’ uses all A* spaces while ‘fast’ skips A2 (3D conformation) which is very slow. 
#                                False by default, not adding any signature to the universe. 
#    sign2_coverage (string): path to the coverage of signatures 2 for all molecules in the CC universe 
#    overwrite (boolean): overwriting the existing signature 
#    validations (boolean): set to False to skip post-signature data validations (including diagnostics which is run separately in a later step)
#                           Current validation tests are performed on MOA (Mode Of Action) and ATC (Anatomical Therapeutic Chemical) corresponding to B1.001 and E1.001 space datasets.
#    complete_universe: it completes the universe for extra molecules. Important if the dataset we are fitting is defined on molecules that largely do not overlap with CC molecules. 
#                       In that case there is no orthogonal information to derive sign3 so we should always have at least 
#                       the chemistry (calculated) level available for all molecules of the dataset.
#    cpu (int): HPC parameter, by default cpu=1
sign3.fit_hpc(sign2_list, sign2_self, sign1_self, None, False, None, cpu=8)
```

```python
# Producing sign3 (full) diagnostic plots
sign3 = cc_local.signature(dataset, 'sign3')
# diagnosis parameter: 
#    ref_cctype (string): the reference for the diagnostics is, by default, sign0 of the same dataset 
diag3 = sign3.diagnosis()
# including the current dataset (M1.001) into the ROC diagnostic plot: by default just the official 25 CC spaces are used 
include_datasets = ['M1.001']
diag3.across_roc(include_datasets=include_datasets, save=True, plot=False) 
```

```python
# Diagnostic canvas parameters: 
#     size (string): canvas size can be 'medium' (default) or 'small' (including a basic version of diagnostics) 
#     savefig (boolean): set to True to save canvas in 'diags' folder 
diag3.canvas(size='medium', savefig=True)
```
