```python
import sys
import os
import json
import h5py
import tqdm
import numpy as np
import pandas as pd
from rdkit import Chem
import matplotlib.pyplot as plt
from collections import Counter
```

```python
########################################################
### PREPARE MAPPINGS: FROM COMPOUND NAME TO INCHIKEY ###
########################################################

# TRANSLATE COMPOUNDS [NAME] TO INCHIKEYS #
cpds_to_ik = pd.read_excel("../data/D6/TableS2.xlsx")

# Caution; "Pirenzepine" is duplicated ...
cpds_to_smiles = {i: j for i, j in zip(cpds_to_ik['Compound Name'], cpds_to_ik['SMILES'])}

# Fix a couple of corner cases
cpds_to_smiles['BACE1'] = cpds_to_smiles['BACE1_In']
cpds_to_smiles['T-00127'] = cpds_to_smiles['T-00127_HEV1']

# SMILES to IKs; Tazarotene1 and Tazarotene2 have the same IK
cpds_to_ik = {i: Chem.MolToInchiKey(Chem.MolFromSmiles(cpds_to_smiles[i])) for i in sorted(cpds_to_smiles)}
```

```python
# Load data
df = pd.read_csv("../data/D6/TableS1.csv",  encoding='windows-1252')

# Get Uniprots and get indices to sort them
uniprots = df['UniprotID'].to_numpy()
ind_uniprots = np.argsort(uniprots)

# Get compounds [remove 'Gene name' and 'UniprotID']
compounds = df.columns.to_numpy()[2:]
compounds = np.array([cpds_to_ik[i.split("_")[0]] for i in compounds])
ind_compounds = np.argsort(compounds)

# Get matrix [remove Gene names and Uniprot IDs]
M = df.values[:, 2:]

# Transpose matrix [compounds as rows, Uniprot IDs as columns]
M = M.T

# Sort by compounds [rows]
M = M[ind_compounds]
compounds = compounds[ind_compounds]

# Sort by Uniprot IDs [columns]
M = np.array([i[ind_uniprots] for i in M])
uniprots = uniprots[ind_uniprots]

# Create pandas dataframe
M = pd.DataFrame(M, index=compounds, columns=uniprots)

# Adjust this for simplicity
df = df.set_index('UniprotID')
```

```python
M
```

```python
# Check that everything is fine ...
cpds = df.columns[1:]
for cpd in tqdm.tqdm(cpds):
    if cpd not in ['Pirenzepine_CHRM1', 'Pirenzepine_CHRM1.1', 'Tazarotene1_RARB', 
                   'Tazarotene2_RARB']:  # Compounds with same Inchikey, leading to a duplicate result...
        for uni in uniprots:
            val1 = df[cpd][uni]
            val2 = M[uni][cpds_to_ik[cpd.split("_")[0]]]
            if (val1 != val2):
                if np.isnan(val1) == False or np.isnan(val2) == False:
                    print(cpd, uni, val1, val2, "Error when processing raw data")
```

```python
# Get values
values = M.values.flatten().astype(float)
not_nan_values = values[~np.isnan(values)]

print("Non nan values: " + str(round(len(not_nan_values) / len(values) * 100, 1)) + " %")

# View histogram
plt.figure(figsize=(4, 3))
plt.hist(not_nan_values, bins=[i for i in np.arange(-2, 2, 0.05)], ec="k")
# plt.yscale('log')
plt.show()
```
