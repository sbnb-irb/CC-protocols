```python
import os
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import pickle
import h5py
from collections import Counter
from sklearn.manifold import TSNE
import random
from sklearn.metrics import pairwise_distances
import tqdm
from scipy.spatial import distance
from scipy.interpolate import interp1d
from sklearn import metrics
import pandas as pd
import sys
import scipy.stats as ss
```

```python
PATH_TO_CC = "/aloy/web_checker/package_cc/2021_07/full/"
# PATH_TO_CC = "/aloy/web_checker/current/full/"
spaces = [i + j for i in "ABCDE" for j in "12345"]
space_to_color = {'A': "#EA5A49", 'B': "#B16BA8", 'C': "#5A72B5", 'D': "#7CAF2A", 'E': "#F39426"}
# space_to_spacefull = {i: i + ".001" for i in spaces}
OUTPATH = "/aloy/home/acomajuncosa/Protocols/rebutal_analyses/applicability/"
```

```python
def get_applicabilities(path_to_h5):
    with h5py.File(path_to_h5) as h5:
        keys = np.array([i.decode('utf-8') for i in h5['keys'][:]])
        apps = h5['applicability'][:] 
    return {i:j for i,j in zip(keys, apps)}
```

```python
#####################################
#####################################
##########     sign3     ############
#####################################
#####################################
```

```python
APPLICABILITIES = {}

# For each space
for space in tqdm.tqdm(spaces):

    # Get the applicabilities
    PATH = os.path.join(PATH_TO_CC, space[0], space, space + ".001", "sign3", "sign3.h5")
    app = get_applicabilities(PATH)
    APPLICABILITIES[space] = app
```

```python
fig, ax = plt.subplots(5, 5, figsize=(7, 8), sharex=True, sharey=True)
letter_to_ind = {i: int(j)-1 for i,j in zip("ABCDE", "12345")}
spaces_to_ind = {i: [letter_to_ind[i[0]], int(i[1])-1] for i in spaces}

for space in spaces:

    # Get c1;c2
    c1, c2 = spaces_to_ind[space]

    # Adjust limits
    ax[c1][c2].set_xlim([0, 1])
    ax[c1][c2].set_ylim([0, 4])
    ax[c1][c2].set_yticks([])
    ax[c1][c2].grid(linestyle='--', zorder=-2)
    ax[c1][c2].set_title(space, pad=10)
    ax[c1][c2].hist(np.array(list(APPLICABILITIES[space].values())), zorder=2,
                    bins=[i for i in np.arange(0, 1, 0.05)], density=True, ec='k', lw=0.5, color=space_to_color[space[0]])

    # xticks
    if space[0] == "E":
        ax[c1][c2].set_xticks([0.25, 0.50, 0.75])
        ax[c1][c2].set_xticklabels(["0.25", "0.50", "0.75"], rotation=90)
        ax[c1][c2].set_xlabel("App. Score", labelpad=10)

    # yticks
    if space[1] == "1":
        ax[c1][c2].set_ylabel("Density", labelpad=10)

plt.tight_layout()
plt.savefig(os.path.join(OUTPATH, 'sign3.png'), dpi=300)
plt.show()
```

```python
[[i[0], i[1], CORRELATIONS[i].statistic] for i in CORRELATIONS if CORRELATIONS[i].statistic > 0.1]
```
