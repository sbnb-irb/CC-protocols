```python
# Read space
space = "A5"

# Some input variables
n_samples = 5
n_mols = 100
aurocs = []
PATH_TO_OUTPUT = "/aloy/home/acomajuncosa/Protocols/rebutal_analyses/information_loss/"
PATH_TO_PVALUES = os.path.join(PATH_TO_OUTPUT, 'pvalues')
PATH_TO_AUROCS = os.path.join(PATH_TO_OUTPUT, 'aurocs')
PATH_TO_CC = "/aloy/web_checker/package_cc/2021_07/full/"
combinations = ["sign0-sign1", "sign0-sign2", "sign0-sign3", "sign0-sign4", "sign1-sign2", 
                "sign1-sign3", "sign1-sign4", "sign2-sign3", "sign2-sign4", "sign3-sign4"]

combinations = ["sign3-sign4"]



# Set random seed
random.seed(0)

for comb in combinations:

    # Specify signature types
    s1, s2 = comb.split("-")

    # Load the pvalues
    pvalues = pickle.load(open(os.path.join(PATH_TO_PVALUES, space + "_pvalues.pkl"), "rb"))

    # Get shared molecules between both sign types
    with h5py.File(os.path.join(PATH_TO_CC, space[0], space, space + ".001", s1, s1 + ".h5")) as h5:
        keys_1 = np.array([i.decode('utf-8') for i in h5['keys'][:]])
    with h5py.File(os.path.join(PATH_TO_CC, space[0], space, space + ".001", s2, s2 + ".h5")) as h5:
        keys_2 = np.array([i.decode('utf-8') for i in h5['keys'][:]])

    # Molecules to use (are in both spaces/signs)
    mols_to_use = sorted(set(keys_1).intersection(set(keys_2)))

    # Repetition
    for samp in range(n_samples):

        # Subsampling
        if len(mols_to_use) > n_mols:
            mols_to_use_SUBSAMPLE = random.sample(sorted(mols_to_use), n_mols)
        else:
            mols_to_use_SUBSAMPLE = mols_to_use

        # Get indices
        ind_1 = np.where(np.isin(keys_1, mols_to_use_SUBSAMPLE))
        ind_2 = np.where(np.isin(keys_2, mols_to_use_SUBSAMPLE))

        # Get signatures
        with h5py.File(os.path.join(PATH_TO_CC, space[0], space, space + ".001", s1, s1 + ".h5")) as h5:
            X_1 = np.array(h5['V'][ind_1])
        with h5py.File(os.path.join(PATH_TO_CC, space[0], space, space + ".001", s2, s2 + ".h5")) as h5:
            X_2 = np.array(h5['V'][ind_2])

        # Calculate distances
        y_pred_1, y_pred_0 = [], []
        for c, (i1, j1) in enumerate(zip(X_1, X_2)):
            for i2, j2 in zip(X_1[c+1:], X_2[c+1:]):
                dist_1 = distance.cosine(i1, i2)
                dist_2 = distance.cosine(j1, j2)
                if dist_1 <= np.mean(pvalues['0.01']):
                    y_pred_1.append(-dist_2)
                else:
                    y_pred_0.append(-dist_2)

        # Calculate AUROC
        y_pred = y_pred_1 + y_pred_0
        y_t = [1] * len(y_pred_1) + [0] * len(y_pred_0)
        fpr, tpr, _ = metrics.roc_curve(y_t, y_pred)
        auroc = metrics.roc_auc_score(y_t, y_pred)

        # Save results
        aurocs.append([space, comb, n_mols, samp, round(auroc, 5)])

        sys.stderr.write(space + " -- " + comb + " -- " + str(samp) + "\n")
        sys.stderr.flush()


# # Save to csv
# aurocs = pd.DataFrame(aurocs, columns=['Space', 'Combination', 'Number of molecules', 'Sample#', 'AUROC'])
# aurocs.to_csv(os.path.join(PATH_TO_AUROCS, space + ".tsv"), sep='\t', index=False)
```
