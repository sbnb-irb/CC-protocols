```python
# Molecules to use
mols_to_use = set(keys_to_sign_D1001).intersection(keys_to_sign_D1002).intersection(keys_to_sign_D6001)
mols_to_use = sorted(mols_to_use)
print(len(mols_to_use))
```

```python
### CALCULATE PVALUES ###

SUBSAMPLE = 5000
random.seed(0)

pvalues = {"D1.001": {}, "D1.002": {}}

### D1.001 ###
pvalues_001, pvalues_00001 = [], []
for r in tqdm.tqdm(range(10)):
    S1 = np.array([keys_to_sign_D1001[i] for i in random.sample(mols_to_use, SUBSAMPLE)])
    S2 = np.array([keys_to_sign_D1001[i] for i in random.sample(mols_to_use, SUBSAMPLE)])
    dist = pairwise_distances(S1, S2, metric = 'cosine', n_jobs=4).flatten()
    pvalues_001.append(np.percentile(dist, 0.01 * 100))  # 10**-2
    pvalues_00001.append(np.percentile(dist, 0.001 * 100))  # 10**-3

pvalues['D1.001']['0.01'] = np.average(pvalues_001)
pvalues['D1.001']['0.001'] = np.average(pvalues_00001)


### D1.002 ###
pvalues_001, pvalues_00001 = [], []
for r in tqdm.tqdm(range(10)):
    S1 = np.array([keys_to_sign_D1002[i] for i in random.sample(mols_to_use, SUBSAMPLE)])
    S2 = np.array([keys_to_sign_D1002[i] for i in random.sample(mols_to_use, SUBSAMPLE)])
    dist = pairwise_distances(S1, S2, metric = 'cosine', n_jobs=4).flatten()
    pvalues_001.append(np.percentile(dist, 0.01 * 100))  # 10**-2
    pvalues_00001.append(np.percentile(dist, 0.001 * 100))  # 10**-3

pvalues['D1.002']['0.01'] = np.average(pvalues_001)
pvalues['D1.002']['0.001'] = np.average(pvalues_00001)

del pvalues_001, pvalues_00001, S1, S2, dist
```
