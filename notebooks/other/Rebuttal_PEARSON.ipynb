```python
### LOAD AND PLOT RESULTS ###
%matplotlib inline

def plot_boxplot(data, position, color, width):
    props = dict(linestyle='-', alpha=1, linewidth=0.8, color='k')
    b = plt.boxplot(data, positions=[position], zorder=2, widths=width, showfliers=False, showcaps=False, medianprops=props, 
                        whiskerprops=props, patch_artist=True, whis=[10, 90])
    b['boxes'][0].set(facecolor = color)
    b['boxes'][0].set(linewidth = 1)
    return b
```

```python
PATH_TO_CC = "/aloy/web_checker/package_cc/2021_07/full/"
PATH_TO_OUTPUT = "/aloy/home/acomajuncosa/Protocols/rebutal_analyses/pearson/results"
spaces = [i + j for i in "ABCDE" for j in "12345"]
space_to_color = {'A': "#EA5A49", 'B': "#B16BA8", 'C': "#5A72B5", 'D': "#7CAF2A", 'E': "#F39426"}
plt.figure(figsize=(7, 3.5))

for c, space in tqdm.tqdm(enumerate(spaces)):

        # Load results
        if os.path.exists(os.path.join(PATH_TO_OUTPUT, space + "_random.pkl")) is True:

            # Pearson CC
            pearson = pickle.load(open(os.path.join(PATH_TO_OUTPUT, space + ".pkl"), "rb"))
            plot_boxplot(np.array(pearson)[:,0], c, color=space_to_color[space[0]], width=0.4)

            # Randomized Pearson CC
            pearson_random = pickle.load(open(os.path.join(PATH_TO_OUTPUT, space + "_random.pkl"), "rb"))
            plot_boxplot(np.array(pearson_random)[:,0], c, color="gray", width=0.4)

plt.title("Colored: type III sign vs. type IV sign.\nGray: type III sign vs. shuffled type IV sign", pad=12)
plt.xlabel("CC Spaces")
plt.ylabel("Pearson CC")
plt.xticks([i for i in range(len(spaces))], spaces, size=9)
plt.grid(linestyle='--', zorder=-2)
plt.tight_layout()
plt.savefig("/aloy/home/acomajuncosa/Protocols/rebutal_analyses/pearson/pearson_cc.png", dpi=300)
plt.show()
```
